```python
import pandas as pd

df_sampah= pd.read_excel('data_sampah_jabar.xlsx',sheet_name='data')
data_sampah=df_sampah[['nama_kabupaten_kota','jumlah_produksi_sampah','tahun']]
data_sampah
data_sampah.to_excel('Data_Sampah.xlsx', index=False)
data_sampah.to_csv('Data_Sampah.csv', index=False)
```

```python
Tahun=2015
sampah2015=0
for i,r in data_sampah.iterrows():
    if r['tahun']==Tahun:
        sampah2015= sampah2015 +r['jumlah_produksi_sampah']
        
print(sampah2015)
df_no2 = pd.DataFrame({'tahun':[Tahun], 'Total produksi sampah':[sampah2015]})
# df_no2
df_no2.to_excel('totalno2.xlsx', index=False)
df_no2.to_csv('totalno2.csv', index=False)
```

```python
data_tahun={}
for i, r in data_sampah.iterrows():
    tahun = r['tahun']
    if tahun not in data_tahun:
        data_tahun[tahun]=0
        
    data_tahun[tahun]+=r['jumlah_produksi_sampah']


print(data_tahun)
df_no3 = pd.DataFrame(list(data_tahun.items()), columns=['tahun', 'jumlah data sampah'])
df_no3.to_excel('jumlah_data_pertahun.xlsx', index=False)
df_no3.to_csv('jumlah_data_pertahun.csv', index=False)
df_no3
```

```python
data_kota={}
for i, r in data_sampah.iterrows():
    tahun = r['tahun']
    kota = r['nama_kabupaten_kota']
    if kota not in data_kota:
        data_kota[kota] = {}    
    if tahun not in data_kota[kota]:
        data_kota[kota][tahun] = 0
    data_kota[kota][tahun] +=r['jumlah_produksi_sampah']
        
# data_kota
df_no4 = pd.DataFrame(data_kota.items(), columns=['kota','tahun'])
df_no4.to_excel('jumlah_data_kota.xlsx', index=False)
df_no4.to_csv('jumlah_data_kota.csv', index=False)
df_no4
```
